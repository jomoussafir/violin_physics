```python
cos_sig = thinkdsp.CosSignal(freq=440, amp=1.0, offset=0)
sin_sig = thinkdsp.SinSignal(freq=880, amp=0.5, offset=0)

mix = cos_sig + sin_sig
wave = mix.make_wave(duration=0.5, start=0, framerate=11025)
```

```python
sin_sig.period*880
```

```python
period = mix.period
print(period)
segment = wave.segment(start=0, duration=period*3)
segment.plot()
thinkdsp.play_wave(segment)
```

```python
wave.make_audio()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
plt.plot(spectrum.fs, abs(spectrum.hs))
```

```python
import matplotlib.pyplot as plt
spectrum.freq_res
spectrum.amps
plt.plot(spectrum.amps)
```

```python
signal = thinkdsp.TriangleSignal(440)
signal.plot()
```

```python
wave = signal.make_wave(duration=1, framerate=10000) 
spectrum = wave.make_spectrum()
spectrum.plot()
```

```python
wave.make_audio()
```

```python
import numpy as np

#x = np.linspace(-10,10,num=1000)
#y = np.sin(x)**8

x = spectrum.fs
y = np.log(spectrum.amps)

plt.figure()
plt.plot(x,y)

dx = np.diff(x, prepend=np.nan)
dy = np.diff(y, prepend=np.nan)
slope = dy/dx
curv = np.diff(slope, append=np.nan)
        
plt.figure()
plt.plot(x, curv, label="dx")
plt.legend()
```
